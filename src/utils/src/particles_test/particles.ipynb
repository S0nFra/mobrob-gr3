```python
import yaml
import numpy as np

def load_yaml(filename):
    data = None
    with open(filename, "r") as stream:
        try:
            data = yaml.safe_load(stream)
        except yaml.YAMLError as exc:
            print(exc)
    return data

def particles_numpy(data):    
    positions = []
    for pose in data['poses']:
        positions.append(list(pose['position'].values()))
    return np.array(positions)
```

```python
data0 = load_yaml("particle02.txt")
print(data0)
```

```python
positions0 = particles_numpy(data0)
print(positions0)
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.scatter(positions0[:,0],positions0[:,1],c='green')
```
